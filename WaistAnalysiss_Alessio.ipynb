```python
import numpy as np
import glob, os

from PIL import Image
import matplotlib.pyplot as plt
from scipy import optimize


#TIFF from RGB to grayscale
def rgb2g(A):

    img = Image.open(A)
    img.getdata()
    r, g, b, k = img.split()

    ra = np.array(r)
    ga = np.array(g)
    ba = np.array(b)

    return((0.299*ra + 0.587*ga + 0.114*ba))

path="/content/drive/MyDrive/Proyectos/Topological transitions of the generalized Pancharatnam-Berry phase/Data/"

J=rgb2g(path+"waist_meas_new/w_1-1.tif")

plt.figure(figsize=(8,8),dpi=100)
plt.imshow(J,cmap='gray')

plt.show()
```

```python
np.shape(J)
```

```python
maxRow=np.argmax(np.max(J, axis=1))

# Fit function for the graph
def Fun_Fit(x, a, b,c):
    return a*np.exp(-(2*(x-b)**2)/(c**2))


x_data=np.arange(1,1281)-1281/2
params, params_covariance = optimize.curve_fit(Fun_Fit, x_data, J[maxRow,:])
waist=params[2]


w=np.array(waist)*0.0052



```
